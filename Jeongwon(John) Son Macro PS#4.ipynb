## 1-a) i.

An infinitely risk-averse private lender will be willing to lend an employed consumer if that consumer is 100% guaranteed to pay his/her interest payments.

If the lender(from here on referred to as agent L) lends assets that are less than $\underline{\textit H}_{-1} \equiv \textit{N}_0/r$, then agent L is guaranteed that the borrower(from here on referred to as agent B) will pay his/her interest payments.

We are assuming that agent B is employed at the time of borrowing, t= -1. Let us think about the case where agent B consumes all of its resources at time t= -1 and then chooses to borrow from agent L the exact amount needed for additional consumption. We refer to this case because it is equivalent to cases where agent B does not consume all of the borrowed resources since the resources that are not consumed will accumulate interest and can be regarded as if these resources were not borrowed in the first place.

If agent B were to become unemployed at t=0, agent B's amount of assets after receiving unemployment benefits will be exactly $\textit {N}_0$, without accounting for debt.

The interest payment($-rA_t$)is smaller than $\textit {N}_0$ since $-A_t < \underline{\textit H}_{-1} \equiv \textit{N}_0/r$. So agent B will be able to pay the interest payment at t=0.

This will hold for all future periods of time as well since agent B receives $\textit {N}_0$ at every period in time after t=0 and interest fees are always smaller than $\textit {N}_0$.

Agent B's income is smallest when agent B becomes unemployed at t=0 since income is growing and becoming unemployed later results in a higher unemployment benefit than $\textit {N}_0$. So, since agent B is able to pay the interest payment continuously when the agent becomes unemployed at t=0, agent B will be able to payoff interest payments in all other cases as well.

Thus, even the most risk-averse lender will choose to lend an amount lower than $\underline{\textit H}_{-1} \equiv \textit{N}_0/r$ since the payment is guaranteed.

## 1-a) ii.

If the consumer has ended employment with $A_{-1}$ amount of assets, at t=0 the consumer's total present discount value of lifetime assets will be $RA_{-1} + \textit {N}_0 + \textit {N}_0/R + \textit {N}_0/R^2 + ...  = RA_{-1} + \textit {N}_0(\frac{1}{1- 1/R}) = RA_{-1} + \textit {N}_0(\frac{R}{r})$.

The consumer will consume strictly positive amounts if this present discount value of lifetime assets is strictly greater than 0. i.e. $RA_{-1} + \textit {N}_0(\frac{R}{r}) > 0$

This is equivalent to $A_{-1} > -\textit {N}_0/r$.

However this is not enough to guarantee a solution to the consumer's optimization problem. The present discount value of total lifetime consumption since t=0 must also be finite.

After unemployment, the consumer's problem is simplified to $$\max_{\{c_t\}} \sum_{t=0}^{\infty} \beta^{t}u(c_{t}) = 1$$
$$\text{s.t.  } m_{t+1} = (m_t - c_t)R + \textit {N}_0$$

This gives us our usual result of consumption growing at $(R\beta)^{1/\rho}$. Thus for the present discount value of total lifetime consumption after unemployment to be finite, we need the condition that consumption growth is less than the discount factor $R$ ($(R\beta)^{1/\rho} < R$).This is equivalent to the 'return impatience condition, $(R\beta)^{1/\rho}/R < 1$.

This will give us the present discount value of total lifetime consumption after unemployment as $c_t + c_{t+1}/R + c_{t+2}/R^2 + ... = c_t + c_t(R\beta)^{1/\rho}/R + ... = c_t(1 + (R\beta)^{1/\rho}/R + ((R\beta)^{1/\rho}/R)^2 + ... = c_t(\frac{1}{1-(R\beta)^{1/\rho}/R})$

## 1-a) iii.

The unemployment benefits program is equal to having a program that provides a lumpsum payment of $\textit {N}_0(R/r)$ upon entering the first period of unemployment since the present discount value of lifetime benefits from the unemployment benefits program is equal to this lumpsum amount.

The present discount value of lifetime benefits from the unemployment benefits program is $\textit {N}_0 + \textit {N}_0/R + \textit {N}_0/R^2 +... = \textit {N}_0(1+ 1/R + 1/R^2 + ... = \textit {N}_0(\frac{1}{1-1/R}) = \textit {N}_0(R/r)$. 

We can see that the present discount value of lifetime benefits is equal to the lumpsum payment of $\textit {N}_0(R/r)$.

The only difference between the two cases will occur if there is a liquidity constraint in the case of receiving continuous benefits. However, since the consumer is at an unemployed state and his/her income is fixed at $\textit {N}_0$ every period, there is no uncertainity involved in the income stream of the consumer. In this case, we can safely say that lenders will freely lend assets such that the consumer can borrow from his/her future benefit income.

Lastly, the consumer decides consumption based on the present discount value of lifetime income or assets and since without liquidity contraints both cases are the same, the consumer will choose his/her path of consumption equally.

## 1-a) v.

The Euler equation for the employed consumer is $\frac{c_{t+1}^e}{c_t^e} = \diamondsuit_{\Gamma}(1+\mho((\frac{c_{t+1}^e}{c_{t+1}^u})^\rho -1))^{1/\rho}$ ...... where $\diamondsuit$ represents "Thorn" used in class and so $\diamondsuit_{\Gamma} \equiv \frac{(R\beta)^{1/\rho}}{\Gamma}$

If we assume $\Delta c^e =0$, then $c_t^e = c_{t+1}^u\Pi$ ...... where $\Pi = (\mho^{-1}(\diamondsuit_{\Gamma}^{-\rho}-1)+1)^{1/\rho}$

Without the unemployment benefits, $c_{t+1}^u = (m_t^e - c_t^e)\mathbb{R}\kappa$ ...... where $\mathbb{R} = \frac{R}{\Gamma}$ and $\kappa = (1-R^{-1}(R\beta)^{1/\rho})$

So the Euler equation becomes 

*  $c_t^e = (m_t^e - c_t^e)\mathbb{R}\kappa \Pi$ or $c_t^e = \frac{\mathbb{R} \kappa \Pi}{1+\mathbb{R} \kappa \Pi}m_t^e$

With unemployment benefits, $c_{t+1}^u = ((m_t^e - c_t^e)\mathbb{R} + \eta(\frac{R}{r}))\kappa$ since human wealth has increased by the equivalent lumpsum amount, $\textit {N}_{t+1}(\frac{R}{r}) = \eta \ell_{t+1} W_{t+1}(\frac{R}{r})$.

* It is easier to first consider the unnormalized version of this equation. $\overline{c_{t+1}^u} = ((\overline{m_t^e} - \overline{c_t^e})R + \eta\ell_{t+1}W_{t+1}(\frac{R}{r}))\kappa$ where variables with an overline represents unnormalized versions of that variable.

So the Euler equation becomes

*  $c_t^e = ((m_t^e - c_t^e)\mathbb{R} + \eta(\frac{R}{r}))\kappa \Pi$ or $c_t^e = \frac{\mathbb{R} \kappa \Pi}{1+\mathbb{R} \kappa \Pi}m_t^e + (\frac{R}{r})\frac{\eta \kappa \Pi}{1+\mathbb{R}\kappa\Pi} = \frac{\mathbb{R} \kappa \Pi}{1+\mathbb{R} \kappa \Pi}(m_t^e + \frac{\eta \Gamma}{r})$

Thus, in the case with unemployment benefits, the $\Delta c =0$ locus is shifted to the left by $\underline{\textit {h}} = \frac{\eta \Gamma}{r}$

* Note here that $\underline{\textit {h}} \neq \frac{\eta}{r}$ which is the value that was stated in the question.

The dynamic budget constraint is $m_{t+1}^e = (m_t^e - c_t^e)\mathbb{R} + 1$. This has no factor involved with unemployment so it is unaffected by unemployment benefits.

Assuming $\Delta m =0$, the dynamic budget constraint becomes

*  $m_t^e = (m_t^e - c_t^e)\mathbb{R} + 1$ or $c_t^e = (1-\mathbb{R}^{-1})m_t^e + \mathbb{R}^{-1}$

An important point here is that the $\Delta c =0$ locus is shifted to the left by $\underline{\textit {h}} = \frac{\eta \Gamma}{r}$ while the $\Delta m =0$ locus is unaffected. The steady state which is the intersection of these two loci has been shifted to the left farther than $\underline{\textit {h}} = \frac{\eta \Gamma}{r}$ because the $\Delta m =0$ locus has a positive slope. This would imply that the consumption function $c(m)$ has shifted to the left by more than $\underline{\textit {h}} = \frac{\eta \Gamma}{r}$.

The target ratio of bank balances become negative when consumption ratio is greater than the asset ratio. Thus we must find the value of $\underline{\textit{h}}$ that shifts the $\Delta c^e =0$ locus such that at the steady state $c_t^e = m_t^e$ is satisfied.

At the steady state such that $c_t^e = m_t^e$ is satisfied, three equations must hold. $\Delta c^e=0$, $\Delta m^e =0$ and $c_t^e = m_t^e$.

First let's solve $\Delta m^e =0$ and $c_t^e = m_t^e$ together.

$m_t^e = (1-\mathbb{R}^{-1})m_t^e + \mathbb{R}^{-1}$ $\Rightarrow$ $m_t^e = c_t^e = 1$

Next, let's input $m_t^e = c_t^e = 1$ into $\Delta c^e=0$ to get a value for $\underline{\textit{h}}$.

$1= \frac{\mathbb{R} \kappa \Pi}{1+\mathbb{R} \kappa \Pi}(1 + \underline{\textit{h}})$ $\Rightarrow$ $\underline{\textit{h}} = \frac{1}{\mathbb{R}\kappa\Pi}$

So, if we were to shift the $\Delta c^e =0$ locus to the left by more than $\frac{1}{\mathbb{R}\kappa\Pi}$, the target ratio of bank balances will be negative.

Another factor that we would have to consider is, that the shift cannot be too great such that the two loci $\Delta c^e=0$ and $\Delta m^e=0$ do not meet. We can guarantee they meet if the y-axis intercept of the $\Delta c^e=0$ is less than the y-axis intercept of $\Delta m^e=0$.

y-axis of $\Delta c^e=0 \Rightarrow \frac{\mathbb{R}\kappa\Pi}{1+\mathbb{R}\kappa\Pi}\underline{\textit{h}}$ 

y-axis of $\Delta m^e=0 \Rightarrow \mathbb{R}^{-1}$ 

$\frac{\mathbb{R}\kappa\Pi}{1+\mathbb{R}\kappa\Pi}\underline{\textit{h}} < \mathbb{R}^{-1} \Rightarrow \underline{\textit{h}} < \frac{\mathbb{R}^{-1}+\kappa\Pi}{\mathbb{R}\kappa\Pi}$ 

So, if the size of the shift $\underline{\textit{h}}$ satisfies $\frac{1}{\mathbb{R}\kappa\Pi} < \underline{\textit{h}} < \frac{\mathbb{R}^{-1}+\kappa\Pi}{\mathbb{R}\kappa\Pi}$, the consumer will be in debt at the steady state.


Below is an example that shows a shift in the $\Delta c^e =0$ locus to the left by exactly $\underline{\textit{h}} = 0.9 \times \frac{1}{\mathbb{R}\kappa\Pi} + 0.1 \times\frac{\mathbb{R}^{-1}+\kappa\Pi}{\mathbb{R}\kappa\Pi}$. 

```python
R=1.03
rho=2.5
u=0.00625
G=1.0025/(1-u)
beta=0.975
k=1-((R*beta)**(1/rho))/R
pi=(u**(-1)*((G**(rho)/(R*beta))-1) +  1)**(1/rho)
m = np.linspace(0,300,num=10,endpoint=True)
x = 0.9/((R/G)*k*pi)+0.1*((G/R) + k*pi)/((R/G)*k*pi)
c1= (1-(G/R))*m + (G/R)
c2= (((R/G)*k*pi)/(1+(R/G)*k*pi))*m
c3= (((R/G)*k*pi)/(1+(R/G)*k*pi))*(m+x)
plt.figure(figsize=(20,5))
plt.plot(m,m,'--y',label ='$c=m$')
plt.plot(m,c1,'b',label ='$\Delta m = 0$' )
plt.plot(m,c2,'r', label ='$\Delta c = 0$ no benefits' )
plt.plot(m,c3,'r', label ='$\Delta c = 0$ w/ benefits' )
plt.xlabel('asset ratio m')
plt.ylabel('consumption ratio c')
plt.ylim(0, 10)
plt.xlim(0,300)
plt.legend()
plt.show()

plt.figure(figsize=(5,3))
plt.plot(m,m,'--y',label ='$c=m$')
plt.plot(m,c1,'b',label ='$\Delta m = 0$' )
plt.plot(m,c3,'r', label ='$\Delta c = 0$ w/ benefits' )
plt.xlabel('asset ratio m')
plt.ylabel('consumption ratio c')
plt.ylim(0.99,1.01)
plt.xlim(0.75,1.1)
plt.legend()
plt.show()
```

## 1-a) vi.

An expansion of unemployment benefits corresponds to a relaxation of a 'natural' or self-imposed borrowing constraint. Without unemployment benefits, consumers are bound to not borrow from their future income since if they become unemployed with 0 assets they will experience infinitely negative utility. However with unemployment benefits, consumers will still have some income in the future even when they become unemployed. Thus, their restrictions on borrowing from the future is relaxed. If consumers borrow a reasonable amount of assets from future income, now they will still be able to consume positive amounts even when they are unemployed.

The phase diagram shifts to the left induced by a temporary decrease in the unemployment risk. However after 6 years, the shift is undone as the unemployment risk returns to its original value.

From this temporary shift, the two loci, $\Delta c = 0$ and $\Delta m = 0$, are affected. The $\Delta c = 0$ locus shifts to the left and while the slope of it also increases. The $\Delta m = 0$ locus is shifted downwards while the slope is increased.

This shift causes consumers who are at their steady state to increase consumption ratio at their given level of asset ratio. This is because with a decreased probability of unemployment, given the same level of asset ratio the consumer feels a weaker motive to save for future consumption. Thus, at the time of the shift, saving rates will dramatically fall since while assets hasn't changed, consumption has risen.

This increase in consumption ratio will cause the consumer's asset ratio to fall. A fall in asset ratio will be followed by a decrease in consumption. During this period, the saving rate will steadily rise as the decrease in consumption ratio is higher than the decrease in asset ratio.

When the unemployment risk is returned to its normal state, consumption ratio will fall drastically since under the original unemployment risk, at the given level of asset ratio the motive to save for future consumption will be stronger. This will cause a sharp rise in saving rates.

The saving rate will then slowly fall until it returns to the steady state level that it was before any kink to the unemployment risk was introduced.

Saving rate is negative because the problem stated that the consumer is a debtor meaning that the consumer will consume more than the assets that he/she posseses, leading to a negative saving rate. This is possible if the unemployment benefit is high enough.

There are two kinks in the saving rate due to the changes in risk of unemployment. The phase diagram shifts to the left induced by a temporary increase in unemployment benefits ($\underline{\eta} \Rightarrow \overline{\eta}$). However after 6 years, the shift is undone as unemployment benefits return to their original value.

From this temporary shift only $\Delta c = 0$ is affected. The $\Delta c = 0$ locus shifts to the left.

This shift causes consumers who are at their steady state to increase consumption ratio at their given level of asset ratio. This is because with higher unemployment benefits the consumer feels a weaker motive to save for future consumption. Thus, at the time of the shift, saving rates will dramatically fall since while assets hasn't changed, consumption has risen.

This increase in consumption ratio will cause the consumer's asset ratio to fall. A fall in asset ratio will be followed by a decrease in consumption. During this period, the saving rate will steadily rise as the decrease in consumption ratio is higher than the decrease in asset ratio.

When belief of income growth is returned to its original level, consumption ratio will fall drastically since the motive to save for future consumption will be stronger. This will cause a sharp rise in saving rates.

The saving rate will then slowly fall until it returns to the steady state level that it was before unemployment benefits was changed ($\underline{\eta}$).

Saving rate is negative because the problem stated that the consumer is a debtor meaning that the consumer will consume more than the assets that he/she posseses, leading to a negative saving rate. This is possible if the original unemployment benefit ($\underline{\eta}$) is high enough.

## 2-a)

$D_s \Rightarrow \underline{D_s} = \frac{1}{2}D_s$

If $D_s$ is reduced to $\underline{D_s}$, $P_s$ will fall to $\underline{P_s} = \frac{1}{2}P_s$. 

The Lucas asset pricing model shows us that the price of an asset is equal to the present discount value of the stream of future dividends the asset produces.

$\underline{P_{s,t}} = \beta\mathbb{E}_t[M_{s,t,t+1}\underline{D_{s,t+1}} + M_{s,t,t+2}\underline{D_{s,t+2}} + M_{s,t,t+3}\underline{D_{s,t+3}}+...]$ $\cdots\cdots\cdots$ where $M_{s,t,t+n}$ is the stochastic discount factor of subprime debt from t+n to t.

$= \beta\mathbb{E}_t[M_{s,t,t+1}\frac{1}{2}D_{s,t+1} + M_{s,t,t+2}\frac{1}{2}D_{s,t+2} + M_{s,t,t+3}\frac{1}{2}D_{s,t+3}+...] $

$= \frac{1}{2}\beta\mathbb{E}_t[M_{s,t,t+1}D_{s,t+1} + M_{s,t,t+2}D_{s,t+2} + M_{s,t,t+3}D_{s,t+3}+...] $

$= \frac{1}{2}P_{s,t}$

Since the price and dividend of subprime debt has fallen equally, the price/dividend ratio does not change. $\frac{P_s}{D_s} = \frac{\underline{P_s}}{\underline{D_s}}$

$\underline{D_s} = \frac{1}{2}D_s$

$\underline{P_s} = \frac{1}{2}P_s$

$\frac{\underline{P_s}}{\underline{D_s}} = \frac{P_s}{D_s}$

A drop in $D_s$ has no effect on other risky assets. The price and dividends of other risky assets are the same as before the drop in $D_s$.

$\underline{D_o} = D_o$

$\underline{P_o} = P_o$

$\frac{\underline{P_o}}{\underline{D_o}} = \frac{P_o}{D_o}$

## 2-b)

$Var(D_s) \Uparrow$

With a modified version of the C-CAPM with no riskfree assets, we can get an answer to this scenario.

$$V(m_t) = \max_{\{c_t\}} u(c_t) + \beta \mathbb{E}_t[V(m_{t+1})]$$
$$s.t. m_{t+1} = (m_t - c_t)\mathbb{R}_{t+1} + y_{t+1}$$
$$\mathbb{R}_{t+1} = \varsigma R_{s,t+1} + (1-\varsigma) R_{o,t+1}$$

If we solve this problem with the FOCs and Envelope theorem, we get
$$0=\mathbb{E}_t[\frac{u'(c_{t+1})}{u'(c_t)}(R_{s,t+1} - R_{o,t+1})]$$

With the same approximation methods used in [EquityPremiumPuzzle], we get
$$ \mathbb{E}_t[R_{s,t+1} - R_{o,t+1}] = \rho Cov(log(\Delta c_{t+1}), R_{s,t+1} - R_{o,t+1}) = \rho Cov(log(\Delta c_{t+1}), R_{o,t+1}) - \rho Cov(log(\Delta c_{t+1}), R_{o,t+1})$$

If $Var(D_s)$ increases while $\mathbb{E}(D_s)$ maintains the same level, subprime debt becomes more risky. The return one would expect to receive will become more stable, implying $Var(R_s) \Uparrow$.

Since it is assumed that other factors are not changed, an increase in $Var(R_s)$ will cause the covariance between consumption growth and $R_s$ to increase since $Cov(\Delta log(c), R_s) = Var(\Delta log(c))\times Var(R_s)\times Corr(\Delta log(c), R_s)$. $\cdots\cdots Cov(\Delta log(c), R_s) \Uparrow$.

Since nothing that affects other risky assets has changed, $R_o$ has not changed.

Then since the modified C-CAPM implies that $ \mathbb{E}_t[R_{s,t+1} - R_{o,t+1}] = \rho Cov(log(\Delta c_{t+1}), R_{o,t+1}) - \rho Cov(log(\Delta c_{t+1}), R_{o,t+1})$, an increase in $Cov(\Delta log(c), R_s)$ will cause an increase in $\mathbb{E}_t[R_{s,t+1}]$. $\cdots\cdots\cdots \mathbb{E}_t[R_{s,t+1}] \Uparrow$.

With average returns higher, prices must fall since the average stream of future dividends is the same. $\cdots\cdots\cdots P_s \Downarrow$

Since the average of dividends has not changed while the price of subprime debts has decreased, the price/dividend ratio has decreased.

$P_s \Downarrow$

$D_s$ remains the same

$\frac{P_s}{D_s} \Downarrow$

Other risky assets have not been affected by this change in $Var(D_s)$ so $P_o$, $D_o$ and $\frac{P_o}{D_o}$ remain the same.

## 2-c)

$\rho \Uparrow$   

From the C-CAPM, we know that $ \mathbb{E}_t[R_{i,t+1}] - R = \rho Cov_t(log(\Delta c_{t+1}), R_{i,t+1}) \cdots i\in\{s,t\}$. An increase in relative risk aversion ($\rho \Uparrow$) will result in an increase in the average rate of return ($\mathbb{E}_t[R_{i,t+1}] \Uparrow$) since $Cov_t(log(\Delta c_{t+1}), R_{i,t+1}) > 0$. This is intuitive since for a consumer with higher relative risk aversion to hold an asset that is procyclical, the rate of return on the asset must be higher. With the average stream of dividends at the same level, an increase in the average rate of return will cause the price of asset i to fall ($P_i \Downarrow$).

This story holds for both subprime debts and other risky assets.

However the magnitude of the change in price is different since $Cov_t(log(\Delta c_{t+1}), R_{s,t+1}) > Cov_t(log(\Delta c_{t+1}), R_{o,t+1})$. Subprime debt prices will rise more than other risky assets and thus the price/dividend ratio for subprime debts will increase more as well.

$P_s \Downarrow P_o \downarrow$

$D_s, D_o$ remain the same

$\frac{P_s}{D_s} \Downarrow, \frac{P_o}{D_o} \downarrow$
